Importing the necessary files

```python
import numpy as np
import matplotlib.pyplot as plt
```

defining all the necessary variables

```python
standard_deviation = 1.3
mean = 1.4
left_limit_of_plot = -10
right_limit_of_plot = 10
n_samples = 100
samples_per_unit = (n_samples-1)/(right_limit_of_plot - left_limit_of_plot)
```

making the x-axis array

```python
x_axis = np.array([i/samples_per_unit - right_limit_of_plot for i in range(n_samples)])
x_axis
```

calculating the y-axis

```python
y_axis = np.exp((-(x_axis - mean)**2)/(2*standard_deviation**2))/(standard_deviation*(2*3.1416)**0.5)
y_axis
```

plotting everything finally

```python
plt.plot(x_axis,y_axis)
plt.show()
```
